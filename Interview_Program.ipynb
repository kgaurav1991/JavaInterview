```python
# Two Array Marge

int ar1[] = {1,2,3};
int ar2[] = {4,5};
int size = ar1.length + ar2.length;
int result[] = new int[size];

for(int i=0; i<ar1.length; i++){
    result[i] = ar1[i];
}
for(int i=0; i<ar2.length; i++){
    result[i+ar1.length] = ar2[i];
}
for(int i=0; i<result.length; i++){
    System.out.print(result[i]+" ");
}

Output: 1 2 3 4 5
```

```python
# Array Second Hightes Number

int ar[] = {1,2,6,7,8,3,4,5};
int size = ar.length;

for(int i=0; i<size; i++){
    for(int j=i+1; j<size; j++){
        if(ar[i] > ar[j]){
            int temp = ar[i];
            ar[i] = ar[j];
            ar[j] = temp;
        }
    }
}
System.out.println(ar[size-2]);

Output: 7

```

```python
# String First Word is Captial Letter

String str = "one two three four";
String word[] = str.split(" ");
StringBuffer result = new StringBuffer();

for(String w : word){
    String s = w.substring(0,1).toUpperCase() + w.substring(1);
    result.append(s + " ");
}
System.out.println(result.toString());

Output: One Two Three Four
```
